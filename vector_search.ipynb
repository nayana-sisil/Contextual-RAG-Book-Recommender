```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
import pandas as pd 

books = pd.read_csv('./dataset/books_cleaned.csv')
books.head()
```

- since  we deal with tagged description we  need to  embed it. therefor let's  create text file first

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
raw_documents  = TextLoader('tagged_description.txt',encoding='utf-8').load()

text_splitter = CharacterTextSplitter(chunk_size=10_000_000, chunk_overlap=0, separator="\n")

documents = text_splitter.split_documents(raw_documents)
```
